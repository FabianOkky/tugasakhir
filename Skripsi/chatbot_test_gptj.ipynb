```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
model_id = "FabianOkky/gptj-physic-modelA"
tokenizerA = AutoTokenizer.from_pretrained("FabianOkky/gptj-physic-modelA")
modelA = AutoModelForCausalLM.from_pretrained(
    "FabianOkky/gptj-physic-modelA",
    device_map="auto",          # otomatis ke GPU
    quantization_config=bnb_config          # ini yang penting agar bisa jalan di Colab
)
```

```python
model_id = "FabianOkky/gptj-physic-modelB"
tokenizerB = AutoTokenizer.from_pretrained("FabianOkky/gptj-physic-modelB")
modelB = AutoModelForCausalLM.from_pretrained(
    "FabianOkky/gptj-physic-modelB",
    device_map="auto",          # otomatis ke GPU
    quantization_config=bnb_config          # ini yang penting agar bisa jalan di Colab
)
```

```python
def ask(model, tokenizer, question, pertanyaan, tanya, model_id=None):
    if tanya:
        prompt = f"<|system|>\nKamu adalah tutor fisika SMP kelas 7 yang ramah dan sabar. Evaluasi jawaban siswa berdasarkan konteks pertanyaan berikut.\n<|pertanyaan|>\n{pertanyaan}\n<|user|>\n{question}\n<|assistant|>\n"
    else:
        if model_id == "modelA":
            prompt = f"<|system|>\nKamu adalah tutor fisika SMP kelas 7 yang ramah dan sabar. Jawablah pertanyaan user dengan jelas dan, jika sesuai, berikan pertanyaan lanjutan sederhana untuk mendorong pemikiran siswa.\n<|user|>\n{question}\n<|assistant|>\n"
        else:
            prompt = (
                "<|system|>\n"
                "Kamu adalah tutor fisika SMP kelas 7 yang bertugas menggali pemahaman siswa dengan bertanya reflektif. "
                "Tugasmu hanya bertanya. Jangan memberikan jawaban. Jawabanmu harus berupa pertanyaan yang membuat siswa berpikir.\n"
                f"<|user|>\n{question}\n"
                "<|assistant|>\n"
            )

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    outputs = model.generate(
      **inputs,
      max_new_tokens=80,  # batas lebih realistis, bisa disesuaikan
      pad_token_id=tokenizer.eos_token_id,
      eos_token_id=tokenizer.eos_token_id,  # tetap pakai token akhir
      do_sample=True,
      temperature=0.2 if (model_id == "modelA" and tanya) else (0.3 if model_id == "modelB" else 0.4),
      top_p=0.9 if model_id == "modelA" else 0.6,
      repetition_penalty=1.1,
      num_return_sequences=1
    )

    text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    response = text.split("<|assistant|>")[-1].strip()

    # # Tambahan kontrol keamanan (jika Model B tidak mengeluarkan pertanyaan)
    # if model_id == "modelB" and not response.endswith("?"):
    #     response += " (Ingat, kamu hanya bertugas bertanya reflektif.)"


    return response
```

```python
def loop_chat():
    print("=== Interaksi Chatbot Dimulai ===\nKetik 'exit' untuk keluar.\n")
    modelB_tanya = False
    pertanyaan = ""

    while True:
        user_input = input("👤 Kamu: ")
        if user_input.lower() == "exit":
            print("💬 Chatbot selesai.")
            break

        if user_input.lower() == "tidak tahu" and modelB_tanya:
            model_response = ask(modelA, tokenizerA, user_input, pertanyaan, tanya=True, model_id="modelA")
            print(f"🤖 Model A (Evaluasi): {model_response}")
            modelB_tanya = False
            pertanyaan = ""
            continue

        trigger = random.randint(1, 5)
        if trigger == 3 and not modelB_tanya:
            modelB_question = ask(modelB, tokenizerB, user_input, pertanyaan, tanya=False, model_id="modelB")
            print(f"🎯 Model B: {modelB_question}")
            pertanyaan = modelB_question
            modelB_tanya = True
        else:
            model_response = ask(modelA, tokenizerA, user_input, pertanyaan, tanya=False, model_id="modelA")
            print(f"🤖 Model A: {model_response}")
            modelB_tanya = False
            pertanyaan = ""
```

```python
# Jalankan loop chat
loop_chat()
```
